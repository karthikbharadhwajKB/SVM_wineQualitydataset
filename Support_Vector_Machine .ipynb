```python
def quality_cat(y):
    if y > 6:
        return 1
    elif 4<y<=6:
        return 0
    elif y <= 4:
        return -1
```

```python
y = df_white.quality
```

```python
cat_y = y.apply(quality_cat)
```

```python
cat_y.value_counts()
```

```python
#scaling 
x = df_white.drop(columns='quality',axis=1)
y= cat_y
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler()
sc_x = sc.fit_transform(x)
```

```python
x_sc = pd.DataFrame(sc_x)
```
